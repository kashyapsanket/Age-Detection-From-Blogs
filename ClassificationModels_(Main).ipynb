```
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
```

```
df = pd.read_csv('drive/My Drive/normalizedfeatures.csv')
df.head
```

```
#df['Unnamed: 0']
df.drop(['Unnamed: 0'], axis=1)
df.shape
```

```
df = df.drop(df.columns[0], axis=1)
df.head
df.shape
```

```
temp_df = pd.read_csv('drive/My Drive/BlogData.csv')
labels = temp_df['Age']
```

```
len(labels)
```

```
df.shape
```

```
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils

# encode class values as integers
encoder = LabelEncoder()
encoder.fit(labels)
encoded_Y = encoder.transform(labels)
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(encoded_Y)
```

```
model = Sequential()
model.add(Dense(64, input_dim=88, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(3, activation='softmax'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Fit the model
```

```
model.summary()
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, dummy_y, test_size=0.3, random_state=42)
```

```
model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs=250, batch_size=2048)
```

```
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import confusion_matrix

def ensembleclass(Xtrain, Ytrain, Xtest):
    print('Ho jaaye shuru')
    log_clf = LogisticRegression()
    rnd_clf = RandomForestClassifier()
    svm_clf = LinearSVC()

    voting_clf = VotingClassifier(estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf)],voting='hard')
    voting_clf.fit(Xtrain, Ytrain)
    Y_test = voting_clf.predict(Xtest)
    print("Please work")
    return Y_test
```

```
test_Y = ensembleclass(X_train, y_train1, X_test)
```

```
def accuracy(x, y):
    print("Does it work?")
    counter = 0
    for i in range(len(x)):
        if x[i] == y[i]:
            counter = counter + 1
    print("We shall see")
    return (counter/len(x))*100
```

```
print(len(y_test1), len(test_Y))
acc = accuracy(list(y_test1), list(test_Y))
print(acc)
print(confusion_matrix(list(y_test1), list(test_Y)))
```

```
def linear_svm(Xtrain, Ytrain, Xtest):
    rf = LinearSVC(max_iter=10000)
    rf.fit(Xtrain, Ytrain)
    Y_test = rf.predict(Xtest)
    return Y_test
```

```
svm_y = linear_svm(X_train, y_train1, X_test)
print(len(y_test1), len(svm_y))
acc = accuracy(list(y_test1), list(svm_y))
print(acc)
print(confusion_matrix(list(y_test1), list(svm_y)))
```

```
def random_f(Xtrain, Ytrain, Xtest):
    rf = RandomForestClassifier()
    rf.fit(Xtrain, Ytrain)
    Y_test = rf.predict(Xtest)
    return Y_test
```

```
rf_y = random_f(X_train, y_train1, X_test)
print(len(y_test1), len(rf_y))
acc = accuracy(list(y_test1), list(rf_y))
print(acc)
print(confusion_matrix(list(y_test1), list(rf_y)))
```

```
def logistic_r(Xtrain, Ytrain, Xtest):
    rf = LogisticRegression()
    rf.fit(Xtrain, Ytrain)
    Y_test = rf.predict(Xtest)
    return Y_test
```
